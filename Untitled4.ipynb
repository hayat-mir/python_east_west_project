```python
df["Category"].value_counts()
```

```python
df.groupby("Sub-Category")["Profit"].sum().sort_values().plot(kind="bar", title="Profit by Sub-Category", figsize=(8,4))
```

```python
df.groupby("Sub-Category")["Quantity"].sum().sort_values().plot(kind="barh", title="Quantity by Sub-Category", figsize=(8,6))
```

```python
df.groupby("Customer Name")["Sales"].sum().nlargest(15).plot(kind="barh", title="Top 15 Customers by Sales", figsize=(8,6))
```

```python
df["Ship Mode"].value_counts().plot(kind="bar", title="Orders by Ship Mode",figsize=(6,4))
```

```python
df.groupby("Product Name")["Sales"].sum().nlargest(15).plot(kind="barh", title="Top 15 Products by Sales", figsize=(8,6))
```

```python
df.plot(kind="scatter", x="Sales", y="Profit", title="Sales vs Profit", figsize=(6,4), alpha=0.9,color="red")
```

```python
df["Region"].value_counts().plot(kind="pie", autopct="%.2f%%", title="Order Share by Region", ylabel="")
```

```python
df["Ship Mode"].value_counts().plot(kind="pie", autopct="%.1f%%", title="Order Share by Ship Mode", ylabel="")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df.set_index("Order Date").resample("MS")["Order ID"].nunique().plot(title="Monthly Unique Orders")
```
